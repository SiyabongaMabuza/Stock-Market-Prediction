```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period='max')
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
# target is 1 if price wnt up (price tomorrow is higher than close price)
```

```python
# old data may not be relavent, so we well remove data from before 1990
sp500 = sp500.loc['1990-01-01':].copy()
```

```python
sp500
#data has been set up
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

#simple split for training/testing. 
#we train the all rows excluding last 100, then test use last 100 rows for testing
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors =["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

# precission score will tell us the percentage of how many times stock went up after we predicted it would go up
# so percentage of ho many times target == 1, given we predicted it would be 1

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# start is 2500 because a trading year has about 250 days,
# so we are basically taking 10 years of data, and train first model with it
# step is 250 so we train in year intervals
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = [] #list of data frames, where each one is the prediction for each year
    
    for i in range(start, data.shape[0], step): 
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)
    
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# looking at mean close price from the last 
# 2, 5(trading week), 60(trading month/3 months), 250(year), 1000(4 years) trading days
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    # sum of the number of days that the stock price went up
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
#update model by increasing number of estimators and reducing min sample split
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] #use predict_proba
    preds[preds >= .6] = 1 #custom threshold, only predict that the price will go up iff there is a 60% chance (default is 50%)
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors) #use ratios as predictors
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
